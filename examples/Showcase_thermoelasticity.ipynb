```python
import sys
# Absolute path to the cloned repo
repo_path = r'C:\Users\loren\Documents\Repo\InfraPy'

# Add to sys.path
if repo_path not in sys.path:
    sys.path.append(repo_path)

from infrapy.thermoelasticity import lock_in_analysis
import matplotlib.pyplot as plt
import numpy as np
from infrapy import io
```

```python
thermal_data = io.load_ir_data(r'C:\Users\loren\Documents\Repo') # Example thermal data: [frames, height, width]

fs = 1000     # sampling frequency (Hz)
fl = 117.9    # response frequency (Hz)

# Call lock_in_analysis
magnitude, phase = lock_in_analysis(thermal_data, fs, fl)
```

```python
plt.figure()
plt.imshow(magnitude, cmap='magma', vmin=0, vmax=np.percentile(magnitude, 99))
plt.title('Lock-in Magnitude')
plt.colorbar(label='Magnitude')
plt.figure()
plt.imshow(phase, cmap='bwr', vmin=-180, vmax=180)
plt.title('Lock-in Phase')
plt.colorbar(label='Phase (degrees)')
```
